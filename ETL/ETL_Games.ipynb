```python
##  Import Libraries
import ast
import pandas as pd
import fastparquet
import re
```

READING AND TRANSFORMATION OF DATA

```python
#Read and load the JSON file:
#open the JSON file and read it line by line. 

# Read the file output_steam_games and load the data into a dataframe
df_steam_games = pd.read_json(r"..\Extras\output_steam_games.json", lines=True)

# Remove rows with NaN values in all fields
df_steam_games.dropna(how='all', inplace=True)

# Display the dataframe with the removed values.
df_steam_games
```

```python
# Convert missing 'id' records to 0
df_steam_games['id'] = df_steam_games['id'].fillna(0).astype(int)

# Convert 'id' values to integers
df_steam_games['id'] = df_steam_games['id'].astype(int)

# Display the updated DataFrame
df_steam_games
```

This dataset contains 13 columns and remains with 32,135 rows after removing NaN rows. However, it still contains null values in some columns. The columns it contains are as follows:

publisher: Represents the content publisher company.
genres: Indicates the genre of the item (i.e., the game). It is formed by a list of one or more genres per record.
app_name: Denotes the name of the item (i.e., the game).
title: Refers to the title of the item.
url: Provides the URL of the game.
release_date: Specifies the release date of the item in the format “2018-01-04.”
tags: Represents content tags. It is formed by a list of one or more tags per record.
reviews_url: Contains the URL where the review for that game can be found.
specs: Describes specifications for each item. It is a list with one or more strings representing the specifications.
price: Indicates the price of the item.
early_access: Indicates early access with a True/False value.
id: Represents the unique identifier of the content.
developer: Refers to the content developer.

# release_date

```python
def get_year_release(dateR):
    if pd.notna(dateR):
        if re.match(r'^\d{4}-\d{2}-\d{2}$', dateR):
            return dateR.split('-')[0]
    return 'Date no available'

# Create a new column with the year
df_steam_games['release_year'] = df_steam_games['release_date'].apply(get_year_release)

# Remove the 'release_date' column
df_steam_games = df_steam_games.drop('release_date', axis=1)

# Display the first few rows
df_steam_games.head()
```

```python
df_steam_games['release_year'].unique()
```

# Price

```python
# Replace non-numeric values with NaN
df_steam_games['price'] = pd.to_numeric(df_steam_games['price'], errors='coerce')

# Convert the column to float
df_steam_games['price'] = df_steam_games['price'].astype(float)

# Display the first few rows
df_steam_games.head()

#Define a function rep_to_float that replaces missing (NaN) values with 0.0 and converts other values to float. 
def rep_to_float(value):
    if pd.isna(value):
        return 0.0
    try:
        float_value = float(value)
        return float_value
    except:
        return 0.0

df_steam_games['price'] = df_steam_games['price'].apply(rep_to_float)
print(df_steam_games['price'].dtype)


```

# Transformación de las columnas 'publisher', 'app_name', 'title', 'developer'

```python
# Columns to transform
columns_to_transform = ['publisher', 'app_name', 'title', 'developer']

# Fill missing values
df_replacement = df_steam_games[columns_to_transform].fillna('Data not available')

# Drop the original columns and concatenate the filled columns with the entire dataframe
df_steam_games = pd.concat([df_steam_games.drop(columns_to_transform, axis=1), df_replacement], axis=1)

df_steam_games.rename(columns={'title': 'item_name'}, inplace=True)

# Display the first few rows
df_steam_games.head()
```

# Transformación columna 'genres'

```python
# Explode the 'genres' column
df_steam_games = df_steam_games.explode('genres')

# Drop rows with missing values in the 'genres' column
df_steam_games = df_steam_games.dropna(subset=['genres'])

# Display the first few rows
df_steam_games.head()
```

# Columnas 'tags', 'specs', 'url' y 'reviews_url'

```python
# drop the specified columns (‘tags’, ‘specs’, ‘url’, ‘reviews_url’) from the DataFrame.
df_steam_games = df_steam_games.drop(['tags', 'specs', 'url', 'reviews_url'], axis=1)
df_steam_games.columns
```

```python
df_steam_games.rename(columns={'id':'item_id'}, inplace=True)
```

```python
def ddbb_verification(df_steam_games):
    mi_dict = {"name": [], "tipe": [], "no_nulls_%": [], "nulls_%": [], "nulls": []}

    for column in df_steam_games.columns:
        porcentaje_no_nulls = (df_steam_games[column].count() / len(df_steam_games)) * 100
        mi_dict["name"].append(column)
        mi_dict["tipe"].append(df_steam_games[column].apply(type).unique())
        mi_dict["no_nulls_%"].append(round(porcentaje_no_nulls, 2))
        mi_dict["nulls_%"].append(round(100 - porcentaje_no_nulls, 2))
        mi_dict["nulls"].append(df_steam_games[column].isnull().sum())

    df_info = pd.DataFrame(mi_dict)
    return df_info

# Call the function
df_steam_info = ddbb_verification(df_steam_games)
print(df_steam_info)
```
